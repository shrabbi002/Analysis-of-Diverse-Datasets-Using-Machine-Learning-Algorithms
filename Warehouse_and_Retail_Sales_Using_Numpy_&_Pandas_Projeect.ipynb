
# Fill null values in numerical columns with mean
# Fill null values in categorical columns with 'Unknown'

```python
numerical_columns = ['RETAIL SALES', 'RETAIL TRANSFERS', 'WAREHOUSE SALES']
categorical_columns = ['SUPPLIER', 'ITEM DESCRIPTION', 'ITEM TYPE']

for col in numerical_columns:
    df[col].fillna(df[col].mean(), inplace=True)


for col in categorical_columns:
    df[col].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
```

```python
correlation_matrix = np.corrcoef(df['WAREHOUSE SALES'], df['RETAIL SALES'])
print("\nCorrelation matrix between 'WAREHOUSE SALES' and 'RETAIL SALES':")
print(correlation_matrix)
```
