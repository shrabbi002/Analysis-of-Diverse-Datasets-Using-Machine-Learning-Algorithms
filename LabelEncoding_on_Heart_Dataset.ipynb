```python


columns_to_drop = ['Data_Value_Footnote_Symbol', 'Data_Value_Footnote', 'Location 1']

df = df.drop(columns=columns_to_drop)
```

```python
df.columns
```

```python
columns_to_encode = ['Year', 'GeographicLevel', 'DataSource', 'Class', 'Topic', 'Data_Value_Unit', 'Data_Value_Type', 'StratificationCategory1', 'Stratification1', 'StratificationCategory2', 'Stratification2', 'TopicID']


label_encoder = LabelEncoder()


for column in columns_to_encode:
    df[column] = label_encoder.fit_transform(df[column])

print(df)
```

#MinMaxScaler

```python
columns_to_scale = ['Year', 'GeographicLevel', 'DataSource', 'Class', 'Topic', 'Data_Value_Unit', 'Data_Value_Type', 'StratificationCategory1', 'Stratification1', 'StratificationCategory2', 'Stratification2', 'TopicID']

scaler = MinMaxScaler()
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])


print(df)
```

#StandardScaler

```python
columns_to_scale = ['Year', 'GeographicLevel', 'DataSource', 'Class', 'Topic', 'Data_Value_Unit', 'Data_Value_Type', 'StratificationCategory1', 'Stratification1', 'StratificationCategory2', 'Stratification2', 'TopicID']

scaler = StandardScaler()

df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])

print(df)
```
